#### Formatting the Columns Data 
Here we clean the text data in the **artist**, **song**, and **genre** columns by:  
- Converting all text to title case (first letter capital) using **.str.title()**  
- Removing extra spaces from the beginning and end of the text using **.str.strip()**  

The **head(20)** function  displays the first 20 rows of the updated dataset.

```python
spotify[["artist", "song", "genre"]] = spotify[["artist", "song", "genre"]].apply(lambda x: x.str.title().str.strip())
spotify.head(20)
```

#### Converting Duration to Minutes  
The **duration_ms** column, which contains song duration in milliseconds, is converted to minutes by dividing it by 60,000.  
A new column **duration_min** is created to store the duration in minutes.  
The original **duration_ms** column is then removed using **.drop()** with **inplace=True** to update the dataset directly.  
**.head()** displays the first 5 rows of the updated dataset.

```python
# Convert duration to minutes
spotify['duration_min'] = spotify['duration_ms'] / 60000
spotify.drop(columns=['duration_ms'], inplace=True)
spotify.head()
```

#### Encoding 'explicit' Column  
The **explicit** column, which contains boolean values (**True** or **False**), is encoded into numerical values (**1** for True and **0** for False) using **LabelEncoder**.  
This is done to convert categorical data into a numerical format suitable for analysis and visualization.
**.head()** displays the first 5 rows of the updated dataset.

```python
# Encode 'explicit' (True/False to 1/0)
le_explicit = LabelEncoder()
spotify['explicit'] = le_explicit.fit_transform(spotify['explicit'])
spotify.head()
```

### Identifying Maximum and Minimum Values Across Numerical Columns  

In this step, we loop through important numerical columns and:  
- **.max()**: Finds the maximum value for each column.  
- **.min()**: Finds the minimum value for each column.  
- **spotify[spotify[col] == max_val]**: Filters and displays the rows where the column has its maximum value.  
- **spotify[spotify[col] == min_val]**: Filters and displays the rows where the column has its minimum value.  

This helps to identify which songs have extreme values (highest and lowest) for features like popularity, danceability, energy, tempo, etc.


```python
columns = ['year', 'popularity', 'danceability',
           'energy', 'key', 'loudness', 'speechiness', 'acousticness',
           'instrumentalness', 'liveness', 'valence', 'tempo',
           'duration_min']
for col in columns:
    max_val = spotify[col].max()
    min_val = spotify[col].min()
    print(f"\n{col.upper()}  Max: {max_val}")
    print(spotify[spotify[col] == max_val][['artist', 'song', col]])
    print(f"\n{col.upper()}  Min: {min_val}")
    print(spotify[spotify[col] == min_val][['artist', 'song', col]])
    print("=" * 50)
```

### Popularity Distribution by Top 10 Genres

This boxplot visualizes how song popularity is distributed across the top 10 most frequent genres in the dataset.

#####  Functions we have used
- **spotify['genre'].value_counts().nlargest(10).index**: Identifies the top 10 genres based on the number of songs.
- **spotify[spotify['genre'].isin(top_genre)]**: Filters the dataset to include only songs from those top genres.
- **sns.boxplot(x='genre', y='popularity', data=spotify_top)**: Creates a boxplot to show the popularity distribution for each genre.
  - x='genre': Genres are plotted along the x-axis.
  - y='popularity': Popularity scores are plotted on the y-axis.
  - data=spotify_top: Uses the filtered dataset.
- **plt.xticks(rotation=90)**: Rotates x-axis labels to prevent overlap.
- **plt.title("Popularity by Top 10 Genres")**: Adds a chart title.
- **plt.show()**: Displays the visualization.

```python
# Get Top 10 Genres
top_genre = spotify['genre'].value_counts().nlargest(10).index
# Filter the dataset for top 10 genres
spotify_top = spotify[spotify['genre'].isin(top_genre)]
# Boxplot for Popularity by Top 10 Genres
plt.figure(figsize=(10, 5))
sns.boxplot(x='genre', y='popularity', data=spotify_top)
plt.xticks(rotation=90)
plt.title("Popularity by Top 10 Genres")
plt.show()
```

#### Insights
- Genres like **Pop**, **Hip Hop, Pop**, and **Pop, R&B** have a relatively high **median popularity**, indicating a strong mainstream presence.
- **Dance/Electronic** and **Rock** genres also show consistent performance, though with slightly wider ranges.
- The **interquartile range (IQR)** for most genres is fairly wide, suggesting variability in song success even within popular genres.
- Outliers (small circles) represent songs that are significantly more or less popular than the majority, highlighting both **hit singles** and **underperformers** within each genre.

## Top 10 Artists by Song Count

This bar chart displays the top 10 artists based on the number of songs they have in the dataset.

#####  Functions we have used
- **spotify['artist'].value_counts().head(10)**: Counts how many times each artist appears and selects the top 10.
- **.reset_index()**: Converts the result into a DataFrame with proper columns.
- **sns.barplot(x='count', y='artist', data=top_artists)**: Creates a horizontal bar plot.
  - x='count': Represents the number of songs.
  - y='artist': Represents the artist names.
  - data=top_artists: Uses the filtered top 10 artists DataFrame.
- **plt.grid(True)**: Adds grid lines for visual clarity.
- **plt.title("Top 10 Artists by Song Count")**: Adds a title to the chart.
- **plt.show()**: Displays the plot.

```python
top_artists = spotify['artist'].value_counts().head(10).reset_index()
top_artists.columns = ['artist', 'count']
plt.figure(figsize=(10, 5))
sns.barplot(x='count', y='artist', data=top_artists)
plt.title("Top 10 Artists by Song Count")
plt.grid(True)
plt.show()
```

### Insights
- **Rihanna** tops the chart with the highest number of songs in the dataset.
- **Drake**, **Eminem**, and **Calvin Harris** also have strong representation.
- Artists like **Britney Spears**, **David Guetta**, **Kanye West**, and **Taylor Swift** reflect consistent popularity and frequent collaborations.
- This plot helps identify the most prominently featured artists, which may correlate with trends, commercial success, or playlist placements.

## Top Genres by Count

This bar chart illustrates the top 10 most frequent genres in the dataset based on the number of songs associated with each genre.

#####  Functions we have used
- **spotify['genre'].value_counts().head(10)**: Counts how many songs belong to each genre and selects the top 10.
- **.reset_index()**: Converts the resulting series into a DataFrame.
- **sns.barplot(x='count', y='genre', data=top_genres)**: Creates a horizontal bar chart.
  - x='count': Number of songs in each genre.
  - y='genre': Genre labels on the y-axis.
  - data=top_genres: Uses the top 10 genres DataFrame.
- **plt.grid(True)**: Adds grid lines to improve visual clarity.
- **plt.title("Top Genres by Count")**: Sets the title of the chart.
- **plt.show()**: Displays the final plot.

```python
top_genres = spotify['genre'].value_counts().head(10).reset_index()
top_genres.columns = ['genre', 'count']
plt.figure(figsize=(10, 5))
sns.barplot(x='count', y='genre', data=top_genres)
plt.title("Top Genres by Count")
plt.grid(True)
plt.show()
```

### Insights
- **Pop** is the most dominant genre by a significant margin, showing its widespread appeal.
- Many genres appear in **combinations**, such as **"Hip Hop, Pop"** and **"Pop, Dance/Electronic"**, highlighting the rise of **genre blending** in modern music.
- Genres like **Hip Hop**, **R&B**, and **Dance/Electronic** are frequently paired with Pop, reflecting popular trends in commercial music production.
- The diversity in genre combinations suggests that artists and producers are increasingly experimenting with **cross-genre styles** to reach broader audiences.
